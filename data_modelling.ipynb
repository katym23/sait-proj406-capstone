# Create Dimension Tables

## Complaint

```python
# remove duplicates
dim_complaint = dim_complaint[~dim_complaint.duplicated(keep='first')]
dim_complaint
```

```python
# add index
dim_complaint['complaint_pk'] = range(1, len(dim_complaint) + 1)
dim_complaint
```

```python
# move index to  beginning
# get a list of columns
cols = list(dim_complaint)
# move the column to head of list using index, pop and insert
cols.insert(0, cols.pop(cols.index('complaint_pk')))
# use ix to reorder
dim_complaint = dim_complaint.loc[:, cols]
dim_complaint
```

# Fact Table

Our fact table represents a patient. Imagine the patient walks in and triage information is recorded at the ER by the triage nurse. All other information about the patient (i.e. the other dimension tables) is collected via EHR or other forms at the ER.

```python
data = pd.read_pickle("original_dataset_clean.pkl")
data = data.drop(columns=['Unnamed: 0'])
data
```

Merge complaint

```python
# Specify all columns for merging
merge_columns = dim_complaint.iloc[:, 1:].columns.tolist()  # Add the remaining columns

# Merge dataframes on all the specified columns
fact_patient = pd.merge(data, dim_complaint[['complaint_pk'] + merge_columns], how='left', on=merge_columns)

# Drop unnecessary complaint-related columns from the merged dataframe
columns_to_drop = [col for col in fact_patient.columns if col.startswith('cc_') and col != 'complaint_pk']
fact_patient = fact_patient.drop(columns=columns_to_drop, errors='ignore')

fact_patient
```

Merge demographics

```python
# Specify all columns for merging
merge_columns = dim_demographics.iloc[:, 1:].columns.tolist()  # Add the remaining columns

# Merge dataframes on all the specified columns
fact_patient1 = pd.merge(fact_patient, dim_demographics[['demographic_pk'] + merge_columns], how='left', on=merge_columns)
fact_patient1
```

Merge meds

```python
# Specify all columns for merging
merge_columns = dim_meds.iloc[:, 1:].columns.tolist()  # Add the remaining columns

# Merge dataframes on all the specified columns
fact_patient2 = pd.merge(fact_patient1, dim_meds[['meds_pk'] + merge_columns], how='left', on=merge_columns)
fact_patient2
```

Merge medical history

```python
# Specify all columns for merging
merge_columns = dim_medical_history.iloc[:, 1:].columns.tolist()  # Add the remaining columns

# Merge dataframes on all the specified columns
fact_patient3 = pd.merge(fact_patient2, dim_medical_history[['medical_history_pk'] + merge_columns], how='left', on=merge_columns)
fact_patient3
```

Merge hospital use

```python
# Specify all columns for merging
merge_columns = dim_hospital_use.iloc[:, 1:].columns.tolist()  # Add the remaining columns

# Merge dataframes on all the specified columns
fact_patient4 = pd.merge(fact_patient3, dim_hospital_use[['hosp_use_pk'] + merge_columns], how='left', on=merge_columns)
fact_patient4
```

Merge imaging

```python
# Specify all columns for merging
merge_columns = dim_imaging.iloc[:, 1:].columns.tolist()  # Add the remaining columns

# Merge dataframes on all the specified columns
fact_patient5 = pd.merge(fact_patient4, dim_imaging[['imaging_pk'] + merge_columns], how='left', on=merge_columns)
fact_patient5
```

```python
# Drop unnecessary complaint-related columns from the merged dataframe

columns_to_drop = [col for col in fact_patient5.columns if col not in ['patient_id', 'disposition','dep_name','esi','arrivalmode',
                                                                       'arrivalmonth','arrivalday','arrivalhour_bin',
                                                                       'triage_vital_hr', 'triage_vital_sbp',
                                                                       'triage_vital_dbp', 'triage_vital_rr',
                                                                       'triage_vital_o2', 'triage_vital_o2_device',
                                                                       'triage_vital_temp', 'demographic_pk', 
                                                                       'complaint_pk', 'meds_pk','medical_history_pk', 
                                                                       'hosp_use_pk','imaging_pk']]
fact_patient = fact_patient5.drop(columns=columns_to_drop, errors='ignore')

fact_patient
```
